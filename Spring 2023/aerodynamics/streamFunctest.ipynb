```python
import streamFunc
import numpy as np
import matplotlib.pyplot as plt
```

```python
# set up mesh of x and ys
xs = np.linspace(-3, 3, 100)
ys = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(xs, ys)

# make Vinf
Vinf = 100
```

```python
# calculate function at each point
# make array
vals = np.zeros((len(xs),len(ys)),dtype="float64")
```

```python
### ellipse
# vals = streamFunc.uniform(Y, Vinf)
# vals += streamFunc.sourcesink(-2, 0, X, Y, 10*Vinf)
# vals += streamFunc.sourcesink(2, 0, X, Y, -10*Vinf)

### cylinder
# vals = streamFunc.uniform(Y, Vinf)
# vals += streamFunc.doublet(0, 0, X, Y, 100)

### cylinder with lift and R
# set R
R = 0.5
# add uniform flow
vals = streamFunc.uniform(Y, Vinf)
# add doublet for cylinder w/ kappa=2piVR^2
vals += streamFunc.doublet(0, 0, X, Y, streamFunc.kappa(Vinf, R))
# add vortex with gamma (for 2 stag pts on body it is less than 4piVR)
vals += streamFunc.vortex(0, 0, X, Y, streamFunc.gamma(Vinf, R) - 10)
```

```python
# contourf it
plt.contour(xs, ys, vals, cmap="jet", levels=50)
plt.gca().set_aspect("equal")
plt.colorbar()

# plot circle for help visualizing
# theta = range(0, 360, 1)
# xs = R*np.cos(theta)
# ys = R*np.sin(theta)
# plt.plot(xs, ys)

plt.show()
```
