```python
import PLLT_LiftDistribution as PLLT
import matplotlib.pyplot as plt
import numpy as np
```

```python
# set up airfoil stuff
Airfoil = "8412"    # airfoil
a = 2       # study aoa
a0 = np.degrees(PLLT.TAFT(Airfoil)[0])      # 0 lift aoa
c = 1       # chord
Vinf = 1.56         # free stream velocity
```

```python
# make plot of CL vs AR

# list for vals
ARs = np.arange(1, 11, 1)
CLs = np.empty(len(ARs))

# loop thru and calc
CLs = [PLLT.LiftDistribution(a, a0, b, c, Vinf, b)[2] for b in ARs]

# star data
sCLs = np.array([0.31953997, 0.64578024, 0.72860484])
sARs = np.array([1, 5, 8])

plt.plot(ARs, CLs, label="PLLT")
plt.scatter(sARs, sCLs, label="Star Data", color="red")
plt.title("$C_L$ Data vs PLLT Approximation")
plt.xlabel("Aspect Ratio")
plt.ylabel("$C_L$")
plt.legend()
plt.show()
```

```python
# make plot of CD vs AR

# list for vals
CDs = np.empty(len(ARs))

# loop thru and calc
CDs = [PLLT.LiftDistribution(a, a0, b, c, Vinf, b)[4] for b in ARs]

# star data
sCDs = np.array([0.0643047, 0.054384, 0.0494685])

plt.plot(ARs, CDs, label="PLLT")
plt.scatter(sARs, sCDs, label="Star Data", color="red")
plt.title("$C_{D_i}$ Data vs PLLT Approximation")
plt.xlabel("Aspect Ratio")
plt.ylabel("$C_{D_i}$")
plt.legend()
plt.show()
```
