```python
import matplotlib.pyplot as plt
import matplotlib
import math
import numpy as np
```

```python
epsilon1x = -.0004794
y=0
epsilon2x = .0001345
epsilon3x = .0006449

plt.plot(epsilon3x, y, "o", markersize=10,label="Epsilon 3")
plt.plot(epsilon2x, y, "o", markersize=10,label="Epsilon 2")
plt.plot(epsilon1x, y, "o", markersize=10,label="Epsilon 1")

theta = np.linspace(0, 360, 10000)

mohr1x = .00056215*np.cos(theta) + .00008275
mohr1y = .00056215*np.sin(theta)
plt.scatter(mohr1x,mohr1y,color="black",s=0.5)
mohr2x = .0002552*np.cos(theta) + .0003897
mohr2y = .0002552*np.sin(theta)
plt.scatter(mohr2x,mohr2y,color="black",s=0.5)
mohr3x = .00030695*np.cos(theta) - .00017245
mohr3y = .00030695*np.sin(theta)
plt.scatter(mohr3x,mohr3y,color="black",s=0.5)

plt.axis("scaled")
plt.xlabel("$\epsilon_{n}$")
plt.ylabel("$(1/2)\gamma_{n}$")
plt.title("Mohr's Strain Circles")
plt.legend(loc="upper left")
plt.annotate("(6.449E-4, 0)",xy=(0.0006449,0),xytext=(.00065, .000045))
plt.annotate("(1.345E-4, 0)",xy=(0.0001345,0),xytext=(.00016, .000045))
plt.annotate("(-4.794E-4, 0)",xy=(-0.0004794,0),xytext=(-.00045, .000045))
plt.show()
```
