```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# set up knowns and range to test
vp = np.arange(900, 1455, 5)
T1 = 509.67
y = 1.4
```

```python
# calc shock speed
vs = [(y+1)/4*vg+0.5*np.sqrt((((y+1)/2)**2)*vg**2 + 6871.5*y*T1) for vg in vp]
```

```python
# plot shock speed
plt.plot(vp, vs)
plt.grid()
plt.xlabel("$V_p$ (ft/s)")
plt.ylabel("$V_s$ (ft/s)")
plt.title("Projectile Speed vs Shock Speed")
plt.show()
```

```python
# calculate M1, M2, and Tg
M1 = [v1/(np.sqrt(1717.87*y*T1)) for v1 in vs]
M2 = [np.sqrt((m1**2+2/(y-1))/((2*y)/(y-1)*m1**2 - 1)) for m1 in M1]
Tg = [T1*(1+(y-1)/2*m1**2)/(1+(y-1)/2*m2**2) - 459.67 for m1, m2 in zip(M1, M2)]
```

```python
# plot Tg
plt.plot(vp, Tg)
plt.grid()
plt.xlabel("$V_p$ (ft/s)")
plt.ylabel("$T_g$($^\circ F$)")
plt.title("Static Temp vs Projectile Speed")
plt.show()
```
