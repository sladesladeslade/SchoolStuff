### This code plots line or contour plots of a function when given parameters.

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# print what dis code do
print("This code plots line or contour plots of a function from its parameters.")
```

```python
# function to get and check inputs
def inputs():

    # prompt for inputs
    A = float(input("A ="))
    B = float(input("B ="))
    n = float(input("n ="))
    m = float(input("m ="))
    O = float(input("O ="))
    Y = float(input("Y ="))

    # ask if inputs are correct
    # make flag for correct nums
    correct = False

    # set up loop to check if values are correct
    while correct == False:
        # display inputs
        print("You input: A =",A," B =",B," n =",n," m =",m," O =",O," Y =",Y)

        # query if inputs are right
        ans = input("Is that correct? (Y or N) ")

        # if ans=N, ask for inputs again
        if ans == "N":
            A = float(input("A ="))
            B = float(input("B ="))
            n = float(input("n ="))
            m = float(input("m ="))
            O = float(input("O ="))
            Y = float(input("Y ="))
            
        # if ans=Y, set flag to True
        elif ans == "Y":
            correct = True

    # return correct values
    return [A, B, n, m, O, Y]
```

```python
# function to perform plotting
def plot(inputs):

    # init vars in case you don't want them
    contourname = "none"
    vertname = "none"
    horizname = "none"
    contpath = "none"
    vertpath = "none"
    horizpath = "none"

    # loop to ask for graph type or to stop
    while True:
        
        # ask for type of plot or to end
        ans = int(input("Would you like a contour plot (1), a line plot (2), or to end (3)? "))

        # if ans=1, generate contour plot
        if ans == 1:

            # ask for x and y ranges
            xmin = float(input("Input the minimum val of x: "))
            xmax = float(input("Input the maximum val of x: "))
            ymin = float(input("Input the minimum val of y: "))
            ymax = float(input("Input the maximum val of y: "))

            # create list of points between min and max vals
            xs = np.arange(xmin, xmax, 1)
            ys = np.arange(ymin, ymax, 1)

            # create a grid of x and y values
            [X, Y] = np.meshgrid(xs, ys)

            # calculate f values
            fs = ((inputs[0]*np.sin(inputs[2]*np.pi*X + (inputs[4]))) + (inputs[1]*np.sin(inputs[3]*np.pi*Y + (inputs[5]))))

            # plot contour
            plt.contourf(X, Y, fs, 20, cmap="magma")

            # set plot range
            plt.axis([xmin, xmax, ymin, ymax])

            # add plot formatting
            plt.colorbar()
            plt.title("Contour Plot")
            plt.xlabel("X Axis")
            plt.ylabel("Y Axis")

            # save plot
            contourname = "ContourPlot.jpg"
            plt.savefig(contourname)
            contpath = os.getcwd() + "\ContourPlot.jpg"

            # display plot
            plt.show()
            continue

        # if ans=2, generate line plot
        elif ans == 2:
            # ask which direction line plot
            ans = input("Would you like a vertical (V) or horizontal (H) line plot? ")

            # if V, do this
            if ans == "V":
                # get inputs of range and constant val
                fixedv = float(input("Input the value of x: "))
                vmin = float(input("Input the minimum value of y: "))
                vmax = float(input("Input the maximum value of y: "))

                # create x and y lists
                ys = np.arange(vmin, vmax, 0.0314)
                fs = np.zeros(ys.size)

                # calculate fxn vals
                for i in range(ys.size):
                    fs[i] = ((inputs[0]*np.sin(inputs[2]*np.pi*fixedv + (inputs[4])))
                    + (inputs[1]*np.sin(inputs[3]*np.pi*ys[i] + (inputs[5]))))

                # plot graph
                plt.plot(ys, fs)

                # plot formatting
                plt.title("Vertical Line Plot")
                plt.xlabel("Y Values")
                plt.ylabel("F Values")

                # save plot
                vertname = "VerticalPlot.jpg"
                plt.savefig(vertname)
                vertpath = os.getcwd() + "\VerticalPlot.jpg"

                # display plot
                plt.show()
                continue

            # if H, do this
            if ans == "H":
                # get inputs of range and constant val
                fixedv = float(input("Input the value of y: "))
                vmin = float(input("Input the minimum value of x: "))
                vmax = float(input("Input the maximum value of x: "))

                # create x and y lists
                xs = np.arange(vmin, vmax, 0.0314)
                fs = np.zeros(xs.size)

                # calculate fxn vals
                for i in range(xs.size):
                    fs[i] = ((inputs[0]*np.sin(inputs[2]*np.pi*xs[i] + (inputs[4])))
                    + (inputs[1]*np.sin(inputs[3]*np.pi*fixedv + (inputs[5]))))

                # plot graph
                plt.plot(xs, fs)

                # plot formatting
                plt.title("Horizontal Line Plot")
                plt.xlabel("X Values")
                plt.ylabel("F Values")

                # save plot
                horizname = "HorizontalPlot.jpg"
                plt.savefig(horizname)
                horizpath = os.getcwd() + "\HorizontalPlot.jpg"

                # display plot
                plt.show()
                continue

        # if ans=3, end loop and output filenames of plots
        elif ans == 3:
             # print filepaths
            print("Saved contour plot as ",contourname," in path ",contpath)
            print("Saved vertical plot as ",vertname," in path ",vertpath)
            print("Saved horizontal plot as ",horizname," in path ",horizpath)

            # end
            break

        # query again if invalid input
        else:
            continue
```

```python
testinputs = inputs()
plot(testinputs)
```
