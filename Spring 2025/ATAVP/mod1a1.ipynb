# Problem 1

## Part A
A large transport military aircraft that needs to operate from unimproved fields would most likely use turboprop
engines. As high speed efficiency will likely not be a priority for a low speed transport, a turboprop engine would
provide efficiency at lower Mach numbers, as well as giving high low speed thrust required for short takeoff and
landing. It is also more likely to be debris resistant, as something like a turbofan can suck large amount of FOD into
the engines, but props create a lower pressure ratio and do not suck things off the ground as easily. Furthermore,
debris going towards a turboprop will be unlikely to go into the core where it would cause significant damage, and would
likely make small dents in the propeller blades that would have smaller effects on performance.

## Part B
An aircraft that needs supersonic flight capabilities, as well as high efficiency loiter at high subsonic speeds would
likely use a low-medium bypass ratio turbofan with an afterburner. An afterburning low bypass turbofan should be capable
of providing the thrust output required for supersonic flight. A turbojet may be ideal for supersonic flight, but
including some bypass allows for the engine to behave more like a turbofan, which has high efficiency in high subsonic
speeds. A combined cycle architecture may be used to give higher efficiency in both modes, allowing the engine to almost
fully convert between a turbojet and turbofan for the best possible performance in both scenarios.

## Part C
A large subsonic passenger aircraft will likely use a high bypass ratio turbofan engine. The high bypass ratio will give
incredibly high efficiency at the high subsonic speeds and altitudes typical of a commercial passenger plane. For a long
flight, turbofan engines will provide a good balance of flight speed and fuel burn to efficiently move passengers.

## Part D
A vertical takeoff vehicle would be likely to use electric propulsion. The ability to distribute electric propulsors
across an aircraft allows for it to operate in a vertical mode while also being able to transition into a more efficient
forward flight mode. An aircraft with a vertical takeoff requirement will also need to fly efficiently in non-vertical
flight, and an electric propulsion system would enable it to transition from an efficient vertical flight mode to an
efficient forward flight mode. A hybrid-electric architecture may also be used to increase the capabilities of the
aircraft, with electric motors providing propulsion and a hybrid system assisting the battery in energy storage or
delivery. 

## Part E
A cruise missile would likely use a turbojet engine. Cruise missiles do not require the efficiency provided by a
turbofan engine as they cover shorter distances and are completely expendable. Furthermore, a turbofan engine would
require a larger diameter to account for both the core and bypass air, and are much more complex than a turbojet. The
turbojet is also less costly because of the lower complexity, which is beneficial for a system that is not resuable.

## Part F
A vehicle designed for high supersonic flight would likely use a scramjet. At speeds greater than Mach 5, a scramjet is
required since the losses from slowing air to a subsonic speed for combustion would be high. This disqualifies turbojets
and turbofan engines since they are best for high subsonic flight speeds. Ramjets would be a good choice for lower
supersonic speeds less than Mach 5, but are not suitable for higher Mach numbers since the air would still have to be
brought down to subsonic speeds, incurring higher losses than a scramjet system which can handle supersonic combustion gases.

first simplify and substitue with $F_c$
$$
\begin{align*}
    F_c=\frac{\dot{m}_c}{g_c}\left[\left(1+\frac{\dot{m}_f}{\dot{m}_c}\right)V_{ce}-V_0\right] \\
    \frac{F_c}{\dot{m}_0}=\frac{\dot{m}_c}{\dot{m}_0g_c}\left[\left(1+\frac{\dot{m}_f}{\dot{m}_c}\right)V_{ce}-V_0\right] \\
    \frac{F_c}{\dot{m}_0}=\frac{\dot{m_c}}{\dot{m}_b+\dot{m}_c}\frac{1}{g_c}\left[\left(1+\frac{\dot{m}_f}{\dot{m}_c}\right)V_{ce}-V_0\right] \\
    \frac{F_c}{\dot{m}_0}=\frac{\frac{\dot{m}_c}{\dot{m}_c}}{\frac{\dot{m}_b}{\dot{m}_c}+\frac{\dot{m}_c}{\dot{m}_c}}
    \frac{1}{g_c}\left[\left(1+\frac{\dot{m}_f}{\dot{m}_c}\right)V_{ce}-V_0\right] \\
    \frac{F_c}{\dot{m}_0}=\frac{1}{1+\alpha}\frac{1}{g_c}\left[\left(1+\frac{\dot{m}_f}{\dot{m}_c}\right)V_{ce}-V_0\right] \\
\end{align*}
$$

then do a similar process with $F_b$
$$
\begin{align*}
    \frac{F_b}{\dot{m}_0}=\frac{\dot{m}_b}{\dot{m}_0}\frac{1}{g_c}\left(V_{be}-V_0\right) \\
    \frac{F_b}{\dot{m}_0}=\frac{\dot{m}_b}{\dot{m}_b+\dot{m}_c}\frac{1}{g_c}\left(V_{be}-V_0\right) \\
    \frac{F_b}{\dot{m}_0}=\frac{\frac{\dot{m}_b}{\dot{m}_c}}{\frac{\dot{m}_b}{\dot{m}_c}+\frac{\dot{m}_c}{\dot{m}_c}}
    \frac{1}{g_c}\left(V_{be}-V_0\right) \\
    \frac{F_b}{\dot{m}_0}=\frac{\alpha}{1+\alpha}\frac{1}{g_c}\left(V_{be}-V_0\right) \\
\end{align*}
$$

now combine
$$
\begin{align*}
    \frac{F}{\dot{m}_0}=\frac{F_c}{\dot{m}_0}+\frac{F_b}{\dot{m}_0}=\frac{1}{g_c}\left[\frac{1}{1+\alpha}\left(
    \left(1+\frac{\dot{m}_f}{\dot{m}_c}\right)V_{ce}-V_0\right) + \frac{\alpha}{1+\alpha}\left(V_{be}-V_0\right)\right]
\end{align*}
$$

figure out the coefficient for $V_0$
$$
\begin{align*}
    -\frac{1}{1+\alpha}V_0-\frac{\alpha}{1+\alpha}V_0 \rightarrow \frac{-1-\alpha}{1+\alpha}V_0 \rightarrow
    -\frac{(1+\alpha)}{1+\alpha}V_0=-V_0
\end{align*}
$$

recombine and solve to get the final equation
$$
\begin{align*}
    \fbox{$\frac{F}{\dot{m}_0}=\frac{1}{g_c}\left(\frac{1+\frac{\dot{m}_f}{\dot{m}_c}}{1+\alpha}V_{ce} +
    \frac{\alpha}{1+\alpha}V_{be}-V_0\right)$}
\end{align*}
$$

to solve SFC, we just divide by $m_0$, then split into its component mass flows and divide to get in terms of $\alpha$
similar to the last derivation
$$
\begin{align*}

assuming const. altitude and weight, and steady level flight on a standard day:

```python
import numpy as np
import matplotlib.pyplot as plt


# known values
h = 40000.                  # altitude (ft)
Pstd = 2116.2               # pres. SL std (lbf/ft^2)
pr = 0.1858                 # pres. ratio @ 40k
tr = 0.7519                 # temp. ratio @ 40k
W = 0.9*40000.              # flight weight (lbf)
S = 720.                    # wing area (ft^2)

# HF-1 drag coefficients
HF1Ms = np.array([0.8, 1.2, 1.4, 2.0])
HF1K1s = np.array([0.2, 0.2, 0.25, 0.4])
HF1CD0s = np.array([0.012, 0.02267, 0.028, 0.027])

# range of Mach #  and load factors to study
ns = np.array([1., 4.])
Ms = np.linspace(0.3, 2.0, 100)

# linterp fxn
def linterp(xs, ys, val):
    for i in range(len(xs) - 1):
        if xs[i] <= val <= xs[i + 1]: return ys[i] + (val - xs[i])*(ys[i + 1] - ys[i])/(xs[i + 1] - xs[i])

# for each load factor
for i, n in enumerate(ns):
    # storage arrays
    Ls = np.full(len(Ms), n*W)      # constant since L=W for steady level assumption
    Ds = np.empty_like(Ls)
    LDs = np.empty_like(Ls)
    
    # loop through Mach #s
    for k, M in enumerate(Ms):     
        # calc dynamic pressure (lbf/ft^2)
        q = 1.4/2.*pr*Pstd*M**2
        
        # calculate lift coefficient
        CL = n*W/(q*S)
        
        # determine drag eqtn coefs
        if M <= 0.8:
            # subsonic is all the same
            K1 = HF1K1s[0]
            CD0 = HF1CD0s[0]
        else:
            # interpolate for supersonic/sonic
            K1 = linterp(HF1Ms, HF1K1s, M)
            CD0 = linterp(HF1Ms, HF1CD0s, M)
        
        # calculate the drag coef. from the parabolic polar, then drag
        CD = CD0 + K1*CL**2
        D = CD*q*S
        Ds[k] = D
        LDs[k] = Ls[k]/D
        
    # plotting
    # lift and drag together
    plt.figure()
    plt.plot(Ms, Ls, label="Lift")
    plt.plot(Ms, Ds, label="Drag")
    plt.xlim(0.3, 2.0); plt.ylim(0, n*40000.)
    plt.xticks(np.arange(0.3, 2.01, 0.1))
    plt.xlabel("Mach Number"); plt.ylabel("Force (lbf)")
    plt.title(f"HF-1 Lift and Drag, 90% TOGW, 40kft, n={n:.0f}")
    plt.legend(loc="upper right", framealpha=1); plt.grid(); plt.tight_layout()
    
    # L/D on its own
    plt.figure()
    plt.plot(Ms, LDs, label="L/D")
    plt.xlim(0.3, 2.0); plt.ylim(0, 12)
    plt.xticks(np.arange(0.3, 2.01, 0.1))
    plt.xlabel("Mach Number"); plt.ylabel("L/D Ratio")
    plt.title(f"HF-1 Lift to Drag Ratio, 90% TOGW, 40kft, n={n:.0f}")
    plt.grid(); plt.tight_layout()
```

# Problem 5

assuming constant altitude and weight on a standard day:

```python
# known values
h = 40000.                  # altitude (ft)
Pstd = 2116.2               # pres. SL std (lbf/ft^2)
astd = 1116.                # speed of sound SL std (ft/s)
pr = 0.1858                 # pres. ratio @ 40k
tr = 0.7519                 # temp. ratio @ 40k
a = astd*np.sqrt(tr)        # speed of sound @ 40k (ft/s)
W = 0.9*40000.              # flight weight (lbf)
S = 720.                    # wing area (ft^2)
n = 1.                      # load factor (g's)
M = 0.8                     # Mach number

# HF-1 drag coefficients
HF1Ms = np.array([0.8, 1.2, 1.4, 2.0])
HF1K1s = np.array([0.2, 0.2, 0.25, 0.4])
HF1CD0s = np.array([0.012, 0.02267, 0.028, 0.027])

# range of weight fractions to study
WFs = np.linspace(0.1, 1., 100)

# storage arrays
SLrngs = np.empty_like(WFs)
CCrngs = np.empty_like(SLrngs)

# loop through weight fracs
for i, WF in enumerate(WFs):     
    # calc dynamic pressure (lbf/ft^2)
    q = 1.4/2.*pr*Pstd*M**2
    
    # calculate lift coefficient
    CL = n*W/(q*S)
    
    # determine drag eqtn coefs
    if M <= 0.8:
        # subsonic is all the same
        K1 = HF1K1s[0]
        CD0 = HF1CD0s[0]
    else:
        # interpolate for supersonic/sonic
        K1 = linterp(HF1Ms, HF1K1s, M)
        CD0 = linterp(HF1Ms, HF1CD0s, M)
    
    # calculate the drag coef. from the parabolic polar, then drag
    CD = CD0 + K1*CL**2
    
    # also need to solve for V (ft/s)
    V = M*a
    
    # TSFC from given eqtn for TSFC of HF-1 (lbm/h-lbf)
    TSFC = (1 + 0.35*M)*np.sqrt(tr)

    # range factor (nmi), gc/g0 cancels out
    RFi = CL/CD*V/TSFC*3600/6080
    
    # solve for steady-level range, and CC range
    sSL = 2*(1 - np.sqrt(WF))*RFi
    sCC = np.log(WF)*(-RFi)
    SLrngs[i] = sSL
    CCrngs[i] = sCC
    
    # store the WF=0.5 vals
    if WF == 0.5:
        sSL05 = sSL
        sCC05 = sCC

# plot results
plt.figure()
plt.plot(WFs, SLrngs, label="Steady-Level")
plt.scatter(0.5, sSL05, label=f"S-L rng={sSL05:.0f} nmi @ WF=0.5")
plt.plot(WFs, CCrngs, label="Cruise-Climb")
plt.scatter(0.5, sCC05, label=f"C-C rng={sCC05:.0f} nmi @ WF=0.5")
plt.xlim(0.1, 1.0); plt.ylim(0, 10000.)
plt.xticks(np.arange(0.1, 1.01, 0.1))
plt.xlabel(r"Weight Fraction $\left(W_f/W_i\right)$"); plt.ylabel("Range (nmi)")
plt.title(f"HF-1 Range, 90% TOGW, 40kft, n={n:.0f}")
plt.legend(loc="upper right", framealpha=1); plt.grid(); plt.tight_layout()
```
