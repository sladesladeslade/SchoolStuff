```python
import matplotlib.pyplot as plt
import matplotlib
import math
import numpy as np
```

```python
theta = np.linspace(0, 360, 10000)

P = 22.5
w = .01
h = .015

sigman = (P*(np.cos(theta*math.pi/180))**2)/(w*h)

taun = np.array([[(P*np.cos(theta*math.pi/180) - P*(np.cos(theta*math.pi/180))**3)/(w*h)],
                [0],
                [(-P*((np.cos(theta*math.pi/180))**2)*np.sin(theta*math.pi/180))/(w*h)]])

plt.plot(theta, sigman, label="Normal Stress", linewidth=2.5)
plt.plot(theta, np.linalg.norm(taun), label="Shear Stress", linewidth=2.5)
plt.xlabel("Theta (degrees)")
plt.ylabel("Stress (kPa)")
plt.title("Normal and Shear Stress")
plt.legend(loc="upper right")
plt.xlim(0, 360)
plt.ylim(0)
plt.grid()
plt.show()
```

```python
sigma1x = -120.834
y=0
sigma2x = -6.6697
sigma3x = 53.5037

topx = -33.66515
topy = 87.16885
tau2x = -63.75185
tau2y = 57.08215
tau3x = 23.417
tau3y = 30.0867

plt.plot(sigma3x, y, "o", markersize=10,label="Sigma 1")
plt.plot(sigma2x, y, "o", markersize=10,label="Sigma 2")
plt.plot(sigma1x, y, "o", markersize=10,label="Sigma 3")
plt.plot(topx, topy, "o", markersize=10, color="black", label="Tau 1")
plt.plot(tau2x, tau2y, "o", markersize=10, color="red", label="Tau 2")
plt.plot(tau3x, tau3y, "o", markersize=10, color="yellow", label="Tau 3")

mohr1x = 87.16885*np.cos(theta) - 33.66515
mohr1y = 87.16885*np.sin(theta)
plt.scatter(mohr1x,mohr1y,color="black",s=0.5)

mohr2x = 57.08215*np.cos(theta) - 63.75185
mohr2y = 57.08215*np.sin(theta)
plt.scatter(mohr2x,mohr2y,color="black",s=0.5)

mohr3x = 30.0867*np.cos(theta) + 23.417
mohr3y = 30.0867*np.sin(theta)
plt.scatter(mohr3x,mohr3y,color="black",s=0.5)

plt.axis("scaled")
plt.xlabel("Normal Stress (MPa)")
plt.ylabel("Shear Stress (MPa)")
plt.title("Mohr's Circles")
plt.legend(loc="upper left")
plt.annotate("(53.5037, 0)",xy=(53.5037,0),xytext=(55, 5))
plt.annotate("(-6.6697, 0)",xy=(-6.6697,0),xytext=(-4, 5))
plt.annotate("(-120.834, 0)",xy=(-120.834,0),xytext=(-118, 5))
plt.annotate("(-33.66515, 87.16885)",xy=(-33.66515, 87.16885),xytext=(-30, 89))
plt.annotate("(-63.75185, 57.08215)",xy=(-63.75185, 57.08215),xytext=(-61, 61))
plt.annotate("(23.417, 30.0867)",xy=(23.417, 30.0867),xytext=(25, 34))
plt.show()
```
