First we will find the centroid of the system by finding the centroid of each rectangle and then taking their weighted average. We know that the centroid will be at the exact horizontal midpoint of both rectangles since they share that axis of symmetry.

The top rectangle has a center at $47/2$ so its centroid is at $(0, 47/2 + 3)$. The bottom rectangle has a center at $(0, 3/2)$.

$$
z = \frac{26.5*3*47 + 1.5*3*50}{3*50+3*47} = 13.613
