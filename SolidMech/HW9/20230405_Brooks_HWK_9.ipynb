First we will transform the coordinates of points A and B shown in the sketch below from $\eta, \zeta$ to $y, z$. This was done with fun geometry to give:

$$
A = (-22.1, 60.17) \\

Then we can compute the stress at these points with the equation:
$$
\sigma_x = -\frac{M\cos{\alpha}}{I_y}z - \frac{M\sin{\alpha}}{I_z}y
$$

M can be found by multiplying the force by the length of the bar since the largest moment will be at the wall:
$$
M = (-10\text{kN})(850\text{mm}) = 8500 \text{kNmm}
$$

$$
\sigma_{xA} = -\frac{-8500\cos{339.84}}{1706271}(60.17) - \frac{-8500\sin{339.84}}{498047}(-22.1)
$$

$$
\sigma_{xA} = 0.4114 \text{kN/mm$^2$} = \fbox{$411.4\text{MPa}$} \ \text{(tension)}
$$

$$
\sigma_{xB} = -\frac{-8500\cos{339.84}}{1706271}(-26.62) - \frac{-8500\sin{339.84}}{498047}(31.72)
$$

$$
\sigma_{xB} = -0.3111\text{kN/mm$^2$} = \fbox{$-311.1\text{MPa}$} \ \text{(compression)}
$$

Therefore the maximum tensile stress is 411.1 MPa at A and the maximum compressive stress is 311.1 MPa at B.
