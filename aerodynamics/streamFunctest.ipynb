```python
import streamFunc
import numpy as np
import matplotlib.pyplot as plt
```

```python
# set up mesh of x and ys
xs = np.linspace(-5, 5, 100)
ys = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(xs, ys)

# make Vinf
Vinf = 10
```

```python
# calculate function at each point
# make array
vals = np.zeros((len(xs),len(ys)),dtype="float64")
```

```python
vals = streamFunc.uniform(Y, Vinf)

vals += streamFunc.sourcesink(-2, 0, X, Y, Vinf)
vals += streamFunc.sourcesink(-1, 0, X, Y, 2*Vinf)
vals += streamFunc.sourcesink(0, 0, X, Y, Vinf)
vals += streamFunc.sourcesink(1, 0, X, Y, -3*Vinf)
vals += streamFunc.sourcesink(2, 0, X, Y, -Vinf)
```

```python
# contourf it
plt.contour(xs, ys, vals, cmap="jet", levels=40)
plt.show()
```
